### Anotações em imagens

Aqui vamos anotar uma imagem: desenhar formas e escrever texto sobre ela. É o que usamos para marcar resultados, como a caixa em torno de um objeto detectado e o seu rótulo.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

plt.rcParams.update({'figure.figsize': (5, 5), 'font.size': 12})
print('Setup pronto.')
```

### Trabalhar sobre uma cópia

As funções de desenho do OpenCV alteram a imagem no lugar. Para preservar a original, desenhamos sempre sobre uma cópia.

```python
img = cv2.imread('../images/dog.png')
img_anotada = img.copy()
print('Forma da imagem (altura, largura, canais):', img_anotada.shape)
```

### Desenhar um retângulo

`cv2.rectangle` recebe a imagem, o canto superior esquerdo, o canto inferior direito, a cor em BGR e a espessura da linha.

```python
cv2.rectangle(img_anotada, (180, 100), (690, 650), (0, 255, 0), 5)

plt.imshow(img_anotada[:, :, ::-1])
plt.axis("off")
plt.show()
```

### Linha e círculo
`cv2.line` liga dois pontos e `cv2.circle` recebe o centro e o raio. As cores também são em BGR.

```python
cv2.line(img_anotada, (0, 0), (944, 895), (255, 0, 0), 3)
cv2.circle(img_anotada, (420, 400), 280, (0, 0, 255), 3)

plt.imshow(img_anotada[:, :, ::-1])
plt.axis("off")
plt.show()
```

### Escrever texto

`cv2.putText` recebe a imagem, o texto, a posição inicial, a fonte, a escala, a cor e a
espessura.

```python
fonte = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img_anotada, "Lua", (390, 720), fonte, 1.5, (0, 255, 0), 3)

plt.imshow(img_anotada[:, :, ::-1])
#plt.axis("off")
plt.show()
```

_________________
### Síntese

Sobre uma cópia da imagem, `cv2.rectangle`, `cv2.line`, `cv2.circle` e `cv2.putText`
permitem marcar regiões e escrever rótulos. É a forma usual de exibir o resultado de uma
detecção: uma caixa e um nome sobre a imagem.
